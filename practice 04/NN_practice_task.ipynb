# Задание: Мини-проект по нейронным сетям

В качестве практического задания по NN и последнего задания курса Вам нужно самостоятельно выбрать прикладную задачу, подобрать данные, выбрать архитектуру нейронной сети, обучить модель и оценить качество результата.

Проект можно выполнить в одном из двух форматов:

1. Обучить нейронную сеть с нуля.
2. Дообучить готовую предобученную модель под свою задачу.

Под **«с нуля»** понимается самостоятельное создание архитектуры модели в PyTorch, TensorFlow/Keras или другой библиотеке.

Под **«дообучением»** понимается использование готовой модели, например ResNet, ConvNeXt, BERT, GPT, EfficientNet, ViT, LSTM, GRU и т.д., с адаптацией под выбранную задачу.

## Возможные типы задач

Вы можете выбрать любую задачу, где уместно использовать нейронную сеть.

Примеры задач:

- классификация изображений;
- классификация текстов;
- генерация текста;
- регрессия;
- прогнозирование временных рядов;
- классификация аудио;
- распознавание объектов;
- сегментация изображений;
- анализ медицинских данных;
- анализ финансовых данных;
- анализ геоданных;
- анализ образовательных данных.

## Требования к данным

Вы должны самостоятельно выбрать датасет.

Допускаются:

- открытые датасеты;
- Kaggle;
- Hugging Face Datasets;
- torchvision datasets;
- собственные данные;
- данные из учебных проектов.

Обязательно нужно разделить данные на:

- `train`;
- `validation`;
- `test`.

## Требования к модели

Выберите один из вариантов реализации.

### Вариант 1. Модель с нуля

Примеры архитектур:

- MLP;
- CNN;
- RNN;
- LSTM;
- GRU;
- простая Transformer-like модель.

### Вариант 2. Дообучение готовой модели

Примеры готовых моделей:

- ResNet;
- EfficientNet;
- ConvNeXt;
- MobileNet;
- BERT / RuBERT / DistilBERT;
- GPT / ruGPT;
- ViT;
- Hugging Face model.

## Что должно быть в notebook

Рекомендуемая структура notebook:

1. Постановка задачи.
2. Описание данных.
3. Подготовка данных.
4. Разделение на `train` / `validation` / `test`.
5. Выбор модели.
6. Обучение модели.
7. Графики `loss` / метрик.
8. Оценка на `test set`.
